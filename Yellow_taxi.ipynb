```python
#renaming column
#rename(columns={"OldName": "new_name"})
#inplace=True if you’re okay modifying it directly.
df_clean.rename(columns={'VendorID':'Vendor_id'}, inplace=True)
```

```python
df_clean.dropna() #to drop any null values
```

```python
#sorting data by trip_distance
#inplace function for mutating original data 
df_clean_sorted_trip_dist = df_clean.sort_values(by=['trip_distance'], ascending=False,)
df_clean_sorted_trip_dist.head()
```

```python
# How many of each payment type are represented in the data?
df_clean['payment_type'].value_counts()
```

```python
# What is the average tip for trips paid for with credit card?
avg_cc_tips = df_clean[df_clean['payment_type']==1]['tip_amount'].mean()
print("Avg cc tip: ", avg_cc_tips)
```

```python
# What is the average tip for trips paid for with cash?
avg_cash_tip = df_clean[df_clean['payment_type']==2]['tip_amount'].mean() 
print("Avg cash tip:", avg_cash_tip)
```

```python
#mean on vendor 
df_clean.groupby('Vendor_id').mean(numeric_only=True)
```

```python
# How many times is each vendor ID represented in the data?
df_clean['Vendor_id'].value_counts()
```

```python
# What is the mean total amount for each vendor?
df_clean.groupby(['Vendor_id'])['total_amount'].mean(numeric_only =True)
```
